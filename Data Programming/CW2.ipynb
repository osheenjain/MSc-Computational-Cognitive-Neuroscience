```python
def groupby_sum_total_rename(col_names:list,groupby_col:list, df_name:str):
    """
    This function takes three arguments: 
    col_names: list of column names that need to be summed
    groupby_col: list of column names used to group the dataframe
    df_name: the name of the dataframe that needs to be modified
    """
    #group dataframe by the specified columns and sum the values
    df_group = df.groupby(groupby_col).sum()
    
    #create a new column named 'Total_sum' by adding the values of 'Total' and 'Totals' columns
    df_group = df_group.assign(Total_sum = df_group['Total'] + df_group['Totals'])
    
    #drop the 'Total' and 'Totals' columns
    df_group = df_group.drop(columns=['Total','Totals'])
    
    #rename 'Total_sum' to 'Total'
    df_group = df_group.rename(columns={'Total_sum':'Total'})
    
    #reset the index of the dataframe
    df_group.reset_index(inplace=True)
    
    #update the original dataframe
    globals()[df_name] = df_group
```

#### Function to derive time series data

```python
def reshape_dataframe(file_path,columns_to_drop):
    
    # Read the file and store it in a new dataframe
    time_series = pd.read_csv(file_path)
    
    # Drop certain columns
    time_series = time_series.drop(columns_to_drop, axis=1)
    
    # Convert remaining columns to datetime
    time_series.columns = pd.to_datetime(time_series.columns, format='%Y%m')
    
    # Melt the dataframe to convert columns to rows
    time_series_melted = time_series.melt(value_vars=time_series.columns[:], var_name='columns', value_name='Value')
    time_series_melted = time_series_melted.groupby('columns').sum().reset_index()
    
    # Create a new column with the corresponding values
    time_series_melted = time_series_melted.rename(columns={'columns':'Year'})
    
    # Sort the values
    time_series_melted = time_series_melted.sort_values(by='Year', axis=0)
    
    return time_series_melted
```

#### Function to filter data by each Borough


```python
def groupby_borough(df, borough_name, groupby_col):
    #filter the data for the given borough
    borough_df = df[df['Borough'] == borough_name]
    
    #groupby the given column and sum the total crimes
    borough_group = borough_df.groupby([groupby_col]).sum()
    
    #create a new column named 'Total_sum' by adding the values of 'Total' and 'Totals' columns
    borough_group = borough_group.assign(Total_sum = borough_group['Total'] + borough_group['Totals'])
    
    #drop the 'Total' and 'Totals' columns
    borough_group = borough_group.drop(columns=['Total','Totals'])
    
    #rename 'Total_sum' to 'Total'
    borough_group = borough_group.rename(columns={'Total_sum':'Total'})
    
    #reset the index of the dataframe
    borough_group.reset_index(inplace=True)
    
    return borough_group
```

#### Function to create pie charts of all Boroughs

```python
def create_borough_pie_charts(df, borough_col, major_cat_col):
    boroughs = df[borough_col].unique()
    fig, axs = plt.subplots(len(boroughs), 1, figsize=(20, len(boroughs) * 5))
    for i, borough in enumerate(boroughs):
        borough_df = df.loc[df[borough_col] == borough]
        borough_group = borough_df.groupby(major_cat_col).sum()
        axs[i].pie(borough_group['Total'], labels=None, autopct='%1.1f%%', startangle=90)
        axs[i].legend(borough_group.index, bbox_to_anchor=(1,0.5), loc="center right", bbox_transform=plt.gcf().transFigure)
        axs[i].set_title(f"Total Crimes by Major Category in {borough} Borough")
    plt.show()
```

#### Functions used in Cluster Analysis and Time Series Analysis


```python
def check_stationarity(df, ts):
    rolmean = df.rolling(window=12, center=False).mean()
    rolstd = df.rolling(window=12, center=False).std()
    
    plt.figure(figsize=(10,5))

    plt.plot(df, label='ORIGINAL', alpha=0.80)
    plt.plot(rolmean, label='ROLLING MEAN')
    plt.plot(rolstd, label='ROLLING STD')
    plt.legend(loc='best')
    plt.xlabel('Year')
    plt.ylabel('# of Crimes')
    plt.title('Rolling Mean & Standard Deviation for {}'.format(ts))
    plt.grid()
    plt.show(block = False)
    plt.close()
    
    dftest = adfuller(df, autolag='AIC')

    print('Results of Dickey-Fuller Test:\n')
    print('Test Statistic: {}'.format(dftest[0]))
    print('p-value: {}'.format(dftest[1]))
    print('# Lags Used: {}'.format(dftest[2]))
    print('Number of Observations Used: {}'.format(dftest[3]))
    for key, value in dftest[4].items():
        print('Critical Value ({}): {}'.format(key, value))
```

### 4.2 Checking Stationarity

```python
# Calling the check_stationarity function to get the results of Results of Dickey-Fuller Test for Original Series

check_stationarity(cluster_analysis['ts'], 'Original Time Series')
```

#### Observation

To summarize the results of the Dickey-Fuller test, the **test statistic is -1.918926330256031**, which is not close to -1, indicating that the **London Crime Data is likely non-stationary**. 

The **p-value** is 0.3232106151953819, which is greater than 0.05, indicating that the null hypothesis cannot be rejected and the time series is likely non-stationary.

A non-stationary time series means that the mean, variance, and autocorrelation structure of the crime data change over time. This could be due to a trend in crime rates over time, such as an increase or decrease in crime rates (Covid 19 could be a factor for sudden changes in past couple of years). Or it could be due to seasonality, such as an increase in crime rates during certain months or years.

A non-stationary time series is generally more difficult to model and forecast than a stationary one. Since the mean, variance, and autocorrelation structure change over time, it is harder to identify patterns and trends in the data.

## 5. Cluster Analysis 

I have conducted an analysis to discover similarities or dissimilarities among criminal activity time series from 2010 to 2022 by searching for possible cluster-like structures in the dataset. To achieve this, I applied twp popular clustering algorithms, including [KMeans algorithm](https://en.wikipedia.org/wiki/K-means_clustering) and the 
[Hierarchical Agglomerative algorithm](https://en.wikipedia.org/wiki/Hierarchical_clustering).

For the KMeans which is an unsupervised learning algorithms, I used the [silhouette score](https://en.wikipedia.org/wiki/Silhouette_(clustering)) to determine the correct number of clusters and the right Eps value, respectively. For the Hierarchical Agglomerative algorithm, I empirically determined the number of clusters by cutting the hierarchy at a certain height. Additionally, to obtain a more detailed clustering, I used two different metrics for evaluating the distance among the series, [Euclidean Distance](https://en.wikipedia.org/wiki/Euclidean_distance) and the [Dynamic Time Warping](https://en.wikipedia.org/wiki/Dynamic_time_warping).

### 5.1 By-Borough Series Clustering

The first type of time series that was analyzed is the by-borough time series data. This type of data shows criminal activity in the city, grouped by borough. The dataset contains monthly reports for criminal activity for each of the 33 boroughs, covering the period from 2010 to 2022. The data is not broken down by specific crime categories, instead providing a summary of criminal activity for each borough.

```python
df = pd.read_csv('ts_borough.csv', index_col='date', 
                 parse_dates=True)
```

```python
ts_by_borough_matrix = np.array(
    [df[df['borough'] == b].sort_index()['value'].values 
     for b in df['borough'].unique()])
```

#### Hierarchical Agglomerative Clustering

#### SUMMARY

* **Euclidean Distance**: for the Euclidean Distance I have decided to cut at an height of 10000, hence obtaining 5 
clusters.
* **Dynamic Time Warping**: for the DTW I have deciced to cut at an height of 7500, hence obtaining 4 clusters.

For both the Euclidean Distance and Dynamic Time Warping metrics, it is clear that Westminster, the borough with the highest number of criminal activities, forms a separate cluster. The remaining boroughs are grouped together in a similar way by both metrics.

#### K-Means Clustering

```python
silhouette_scores = []

for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters).fit(ts_by_borough_matrix)
    silhouette_scores.append(silhouette_score(ts_by_borough_matrix, 
                                              kmeans.labels_))
    
plt.figure(figsize=(10, 5))
plt.plot(range(0, 9), silhouette_scores, color='#4bafd1', linewidth=2.5,
         marker='o', markersize=7.0, markevery=[3])
plt.annotate(round(silhouette_scores[3], 2), (3.2, silhouette_scores[2]),
             fontsize=11)
plt.grid()
plt.xticks(ticks=range(0, 9), labels=range(2, 11))
plt.xlabel('# of Clusters', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.title('Silhouette Score per # of Clusters', fontsize=14)
```

#### Overview

Given that the dataset consists of 32 samples, I decided to further analyze the K-Means algorithm by determining the optimal number of clusters to initialize the algorithm with, ranging from 2 to 11. Despite the fact that the value of 5 clusters is not the highest when considering the silhouette score, it provides a more meaningful representation of the relationships among the series in the dataset.

```python
kmeans = pd.DataFrame(data=KMeans(n_clusters=5).fit_predict(ts_by_borough_matrix),
                      index=df['borough'].unique(), columns=['Cluster'])
kmeans.index.name = 'Borough'

kmeans.sort_values('Cluster')
```
